```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Adjust path relative to the notebook's location
with open('../public/generated_motion/results.json') as f:
    data = json.load(f)

motion = np.array(data['motion'][0])  # Sample 0

# Plot every 5th frame
for f in range(0, len(motion), 5):
    joints = np.array(motion[f])
    joints = joints.reshape(-1, 3)  # Ensure it's (N, 3) shape
    xs, ys, zs = joints[:, 0], joints[:, 1], joints[:, 2]

    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(xs, ys, zs)
    plt.title(f'Frame {f}')
    plt.show()
```
